```python
import numpy as np 
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/albanda/CE888/master/lab2%20-%20bootstrap/vehicles.csv')
# print(df.values.shape)
# print(df.head())

print(len(df))

id = np.arange(len(df))
id = id+2
# print(id)

df['id'] = id
print(df.head())


 
```

```python
sns.scatterplot(data=df, x="id", y="Current fleet")
```

```python
sns.scatterplot(data=df, x="id", y="New Fleet")
```

```python
fig, axes = plt.subplots(2,1, figsize=(15, 5), sharey=True)
fig.suptitle('Scatter Plot')
sns.scatterplot(ax=axes[0], data=df, x="id", y="Current fleet")
sns.scatterplot(ax=axes[1], data=df, x="id", y="New Fleet")


```
